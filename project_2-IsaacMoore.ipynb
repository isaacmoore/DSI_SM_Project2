## Step 2: Clean your data.

##### Do some rudimentary cleaning. Rename any columns that are poorly named, shorten any strings that may be too long, fill in any missing values. Explain your rationale for the way you choose to "impute" the missing data.

```python
# Rename columns
data = data.rename(columns=lambda x: x.replace('x', ''))
data = data.rename(columns=lambda x: x.replace('.', '_'))
data.head()
```

```python
# Change NaN(s) to 0
data = data.replace(np.nan,0, regex=True)
data.head()
```

```python
# date_peaked & date_entered - object --> date/time

# data['date_entered'] = pd.to_datetime(data['date_entered'])
data['date_peaked'] = pd.to_datetime(data['date_peaked'])
data.dtypes
```

```python
# Convert genre object category
data['genre'] = data['genre'].astype('category')
data.dtypes
```

```python
# Convert Week 1 from int64 to float
data['1st_week'] = data['1st_week'].astype(float)
data.dtypes
```
